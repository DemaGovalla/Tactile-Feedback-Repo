```python
import gradio as gr
import tensorflow as tf
import numpy as np
import json
from os.path import dirname, realpath, join
```

```python
# load image labels

with open(r"C:\\Users\\dema2\OneDrive\Desktop\PhD\RFMN\Feedback_GUI\imageclassifierapp-main\imagenet_labels.json") as labels_file:
    labels = json.load(labels_file) 

print(labels)
```

```python
# mobile_net = tf.keras.applications.MobileNetV2() # pretrained Convo NN (E.g. After I train the nextwork)

# def image_classifier(im):  # takes in image
#     arr = np.expand_dims(im, axis = 0)
#     arr = tf.keras.application.mobilenet.prepocess_input(arr) # preprocess the image
#     prediction = mobile_net.predict(arr).flatten() # Predict the image
#     return {labels[i]: float(prediction[i]) for i in range(1000)} # retrive labels from json file


mobile_net = tf.keras.applications.MobileNetV2()
def image_classifier(im):
    arr = np.expand_dims(im, axis=0)
    arr = tf.keras.applications.mobilenet.preprocess_input(arr)
    prediction = mobile_net.predict(arr).flatten()
    return {labels[i]: float(prediction[i]) for i in range(1000)}
```

```python
# Build the Web Application using Gradio

iface = gr.Interface(image_classifier,
                     gr.inputs.Image(shape= (224, 224)),
                     gr.outputs.Label(num_top_classes = 5),
                     capture_session = True,
                     interpretation= "default",
                     examples=[[r"C:\Users\dema2\OneDrive\Desktop\PhD\RFMN\Feedback_GUI\imageclassifierapp-main\images\cheetah1.jpg"], 
                               [r"C:\Users\dema2\OneDrive\Desktop\PhD\RFMN\Feedback_GUI\imageclassifierapp-main\images\lion.jpg"]]) 
```
