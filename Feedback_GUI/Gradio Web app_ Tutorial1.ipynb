```python
interface.launch()
```

```python
interface.launch()
```

```python
import numpy as np

def sepia(input_img):
    sepia_filter = np.array([[.393, .769, .189],
                             [.349, .686, .168]
                             [.272, .534, .131]])
    sepia_img = input_img.dot(sepia_filter.T)
    sepia_img /= sepia_img.max()
    return sepia_img

iface = gr.Interface(sepia, gr.inputs.Image(shape=(276,422)), "image")

iface.launch()
```
